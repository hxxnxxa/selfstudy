## ConvNet Example

```python
def convnet_forward(n,filter_size,filters,nc=1,stride=1,padding=0):
    output= ((n+(padding*2)-filters)/stride)+1
    
    return (output,output,filter_size)
    
```

```python
output=convnet_forward(39,10,3,3,1,0)
```

```python
output=convnet_forward(output[0],20,5,output[2],stride=2)
```

```python
output=convnet_forward(output[0],40,5,output[2],stride=2)
```

```python
dimension_vector_size= output[0]*output[1]*output[2]
print("It will create a vector with size: ",int(dimension_vector_size))
```
